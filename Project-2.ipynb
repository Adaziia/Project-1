```python
# converting the categorical value into corresponding numeric representation
data["sex"]=data["sex"].apply({"female":0,"male":1}.get)
```

```python
data["smoker"]=data["smoker"].apply({"yes":1,"no":0}.get)
```

```python
data["region"]=data["region"].apply({"southwest":1,"southeast":2,"northeast":3,"northwest":4}.get)
```

```python
data.iloc[0:10]
snr.relplot(x="sex",y="age",style="smoker",hue="sex",data=data)
```

```python
snr.catplot(x="region",y="smoker",data=data,kind="bar")
```

```python
x=data.drop(["smoker"],axis=1)
y=data["smoker"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier()
```

```python
classifier.fit(x_train,y_train)
```

```python
y_hat=classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,y_hat)
```
