```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier()
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as snr
```

```python
data=pd.read_csv("insurance2.csv")
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
data.iloc[1:10].plot(x="region",y="insuranceclaim",kind="bar")
```

```python
snr.relplot(x="insuranceclaim",y="age",hue="sex",style="smoker",data=data)
```

```python
snr.catplot(x="insuranceclaim",y="age",data=data,hue="sex",kind="box")
```

```python
snr.relplot(x="insuranceclaim",y="age",hue="sex",col="region",style="smoker",data=data)
```

```python
x=data.drop(["insuranceclaim"],axis=1)
y=data["insuranceclaim"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.80)
```

```python
classifier.fit(x_train,y_train)
```

```python
y_hat=classifier.predict(x_test)
```

```python
acc= accuracy_score(y_test,y_hat)
```
